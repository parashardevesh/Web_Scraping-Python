```python
from urllib.request import urlopen as uReq, Request
from bs4 import BeautifulSoup as soup
```

```python
my_url = "https://www.imdb.com/chart/top/?ref_=nv_mv_250"
headers = {'User-Agent': 'Mozilla/0.5'} 
req = Request(my_url, headers=headers)
uClient = uReq(req)
page_html = uClient.read()
uClient.close() 
```

```python
page_soup = soup(page_html, 'html.parser')
#page_soup
```

```python
containers = page_soup.find_all("div", class_ = 'sc-b0691f29-0 jbYPfh cli-children')
print(len(containers))
```

```python
print(soup.prettify(containers[0]))
```

```python
filename = "imdb_m.csv"
f = open(filename, 'w')

headers = "Name, Year, Runtime \n" 
f.write(headers)

for container in containers:
    name= str(container.find('h3', {'class':'ipc-title__text'}))
    year_mov=container.find_all('span', {'class':'sc-b0691f29-8 ilsLEX cli-title-metadata-item'})
    year=year_mov[0].text
    runtime_mov = container.find_all('span', {'class': 'sc-b0691f29-8 ilsLEX cli-title-metadata-item'})
    runtime=runtime_mov[1].text
    
    print(name + ',' + year + ',' + runtime + '\n') 
    f.write(name + ',' + year + ',' + runtime + '\n')

f.close()
```

#let's check our csv file and also edit the name column and remove unnecessary text

```python
import pandas as pd
imdb = pd.read_csv('imdb_m.csv', encoding = 'latin1')
imdb.head()
```

```python
imdb['Name'] = imdb['Name'].apply(lambda x: x.strip('<h3 class="ipc-title__text">'))
imdb.head()
```

```python
imdb.Name = imdb.Name.apply(lambda x: x.strip('</'))
imdb.tail()
```

```python
def replace_func(x):
    x= x.replace('.', '',1)
    for i in range(1,251):
        x= x.replace( str(i), '', 1)
    return x
```

```python
imdb['Name'] = imdb['Name'].apply(replace_func)
```

```python
imdb.head()
```

```python
imdb['Name'] = imdb.Name.apply(lambda x: x.strip('1'))
imdb['Name'] = imdb.Name.apply(lambda x: x.strip('2'))
imdb['Name'] = imdb.Name.apply(lambda x: x.strip('5'))
imdb['Name'] = imdb.Name.apply(lambda x: x.strip('6'))
imdb['Name'] = imdb.Name.apply(lambda x: x.strip('7'))
imdb['Name'] = imdb.Name.apply(lambda x: x.strip('8'))
imdb['Name'] = imdb.Name.apply(lambda x: x.strip('9'))
imdb['Name'] = imdb.Name.apply(lambda x: x.strip('0'))
```

```python
imdb["Name"] = imdb.Name.apply(lambda x: x.replace('0', '',1))
imdb['Name'] = imdb.Name.apply(lambda x: x.replace('3', '',1))
```

```python
imdb['Name'] = imdb.Name.apply(lambda x: x.replace('4', '',1))
```
